### Manipulações com Pandas II

#### Descrição

1. Usando o arquivo "best-selling game consoles.xlsx" conte quantos consoles tem na base por companhias.
2. Ordene o resultado anterior pela quantidade de consoles por companhia de forma decrescente.
3. Sobre o dataframe anterior, adicione a coluna "Market Value" presente na tabela do arquivo "company.txt".
4. Qual o valor médio de Market Value por console?


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('best-selling game consoles.xlsx', sheet_name='consoles')
df.head()
```

Quantos consoles existem por companhia, ordenados de maneira decrescente baseado na quantidade de consoles:

```python
consoles_por_cia = df.groupby('Company')['Console Name'].count().sort_values(ascending=False).reset_index(name='Console Qty')
consoles_por_cia
```

Informação anterior, acrescida do Market Value da companhia:

```python
df_mkt = pd.read_csv('company.txt', sep=',')
consoles_por_cia = consoles_por_cia.merge(df_mkt, on='Company', how='left')
consoles_por_cia
```
