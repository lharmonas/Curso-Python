```python
import pandas as pd
import requests
import schedule
import time
```

```python
url = "https://restcountries.com/v3.1/all"
requisicao = requests.get(url)
if requisicao.status_code == 200:
    data_json = requisicao.json()
else:
    print("Erro ao acessar a API")
```

```python
paises_full = pd.DataFrame(data_json)
```

```python
paises_full.info()
```

```python
nome = [pais["common"] for pais in paises_full["name"]]
region = [pais for pais in paises_full["region"]]
population = [pais for pais in paises_full["population"]]

paises = pd.DataFrame({"Nome": nome, "Regiao": region, "Populacao": population})
paises.head()
```

```python
def buscar_info_uf(sigla):
    url = (f'https://brasilapi.com.br/api/ibge/municipios/v1/{sigla}?providers=gov')
    req_estado = requests.get(url)

    if req_estado.status_code == 200:
        infos_estado = pd.DataFrame(req_estado.json())
        return(infos_estado)
    else:
        print("Erro ao acessar a API")
```

```python
df_AM = buscar_info_uf('AM')
df_AM
```
