```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('realtor-data.csv')
df.info()
```

```python
# df.dropna(subset = ['status', 'city', 'state']) # Formato não automatizado
var_categoria = list(df.columns[df.dtypes == 'object'].values)
df = df.dropna(subset = var_categoria)
df
```

```python
df= df.fillna({"bed":0, "bath":0, "house_size":df["house_size"].median()})
df = df.reset_index(drop=True)
df.head()
```

```python
df['prev_sold_date'] = pd.to_datetime(df['prev_sold_date'])
```

```python
df.dtypes
```

```python
df['prev_ano_venda'] = df['prev_sold_date'].dt.year
df.head()
```

```python
df.pivot_table(index='state', columns='prev_ano_venda', values='price', aggfunc='sum')
```
